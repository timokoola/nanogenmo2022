```python
import os
import openai
import json
import random


openai.api_key = os.getenv("OPENAI_KEY")

chapters = []
prompted_chapters = []
prompt = book_metadata + "\n# " + chapter_one
current_chapter = prompt
chapter_index = 0
word_count = len(prompt.split())

def generate_chapter_start(model):
    chapter_start_prompt = f"Fantasy novel placed in The kingdom of Erasth. The kingdom of Erasth was a peaceful kingdom ruled by a wise and benevolent king. The people lived in harmony with nature and each other, and the kingdom was prosperous and happy. The arrival of the dragon changed all of that. Generate two starting sentences for chapter '{chapter_titles[chapter_index]}'\n\n"
    response = openai.Completion.create(
      model=model,
      prompt=chapter_start_prompt,
      temperature=0.7,
      max_tokens=256,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    )
    prompt = response["choices"][0]["text"]
    return prompt
    
def generate_more(model):
    response = openai.Completion.create(
    model=chapter_model,
    prompt=prompt,
    temperature=1.0,
    max_tokens=1400,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0)
    text = response["choices"][0]["text"]
    return text, response

def next_prompt(text):
    prompt = ". ".join(text.split(".")[-3:]).strip()
    if len(prompt) > 500:
        prompt = ". ".join(text.split(".")[-2:])
    elif len(prompt) == 0:
        prompt = random.choice(sentence_prompts)
    return prompt
    

while word_count < 50000:
    if chapter_index > 0:
        prompt = generate_chapter_start(chapter_model)
    while len(current_chapter.split()) < 1000:
        text, response = generate_more(chapter_model)
        prompted_chapters.append({"prompt": prompt, "text": text, "response": response})
        current_chapter = current_chapter + text
        prompt = next_prompt(text)
        
    chapter_index = chapter_index + 1
    chapters.append(current_chapter)
    current_chapter = "\n\n# " + chapter_titles[chapter_index] + "\n\n"
    prompt = prompt + "\n\n" +  current_chapter
    
    word_count = len("\n".join(chapters).split())
    print(f"word count: {word_count}")
    with open(f"dragons-heart-{chapter_model}.md", "w+") as f:
        f.write("\n".join(chapters))
    with open(f"dragons-heart-{chapter_model}.json", "w+") as f:
        f.write(json.dumps(prompted_chapters, sort_keys=True, indent=4))
    
        
        
    


```
